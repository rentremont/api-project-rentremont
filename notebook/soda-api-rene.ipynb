## 4. Make an API request that returns the entire Top 500 Monthly Searches dataset. Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'


response = requests.get(endpoint)
result = response.json()
result
```

```python
len(result)
```

```python
search_dict = {}
for line in result:
    count = int(line.get('query_count'))
    text = line.get('query_text')
    if text in search_dict:
        search_dict[text] += count
    else:
        search_dict[text] = count
```

```python
search_df = pd.DataFrame.from_dict(search_dict, orient = 'index')
search_df = search_df.reset_index()
search_df.columns = ['term', 'text']
search_df.head()
```

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?query_text=maps'


response = requests.get(endpoint)
result = response.json()
result
```

```python
len(result)
```

```python
map_df = pd.DataFrame.from_dict(result)
map_df
```

```python
map_df['query_count']=map_df['query_count'].astype(int)
```

```python
plt.figure(figsize=(12,15))
plt.bar( 'month_name','query_count', data = map_df)
```

## 6. Find 2 new data sets on data.nashville.gov, make API requests to pull the data, and do an analysis that combines the data sets. 

```python
endpoint = 'https://data.nashville.gov/resource/u64p-38mb.json?calyear=2016'



response = requests.get(endpoint)
result = response.json()
result
```

```python
nes_df = pd.DataFrame.from_dict(result)
nes_df
```

```python
nes_df=nes_df.rename(columns = {'use_tva_cde':'tvacode'})
nes_df
```

```python
endpoint2 = 'https://data.nashville.gov/resource/ut4c-a9ss.json'



response2 = requests.get(endpoint2)
result2 = response2.json()
result2
```

```python
nes2_df = pd.DataFrame.from_dict(result2)
nes2_df
```

```python
nes3_df=pd.merge(nes_df, nes2_df, on=["tvacode"])
```

```python
nes3_df
```

```python
nes3_df.info()
```

```python
nes3_df['revnu_use_kwh']=nes3_df['revnu_use_kwh'].astype(int)
```

```python
plt.figure(figsize=(12,15))
plt.bar( 'calmonth','revnu_use_kwh', data = nes3_df)
```
